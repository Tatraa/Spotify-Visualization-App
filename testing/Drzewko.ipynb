```python
from graphviz import Digraph

tree = Digraph()

tree.node('park_owner','Barbara')

with tree.subgraph() as subgraph:
    subgraph.node('supervisor','Jim Lahey')
    subgraph.node('supervisor2','Randy')
    subgraph.attr(rank='same')

with tree.subgraph() as subgraph2:
    subgraph2.node('junky1','Julian')
    subgraph2.node('junky2','Bubbles')
    subgraph2.node('junky3','Ricky')
    subgraph.attr(rank='same')

with tree.subgraph() as subgraph3:
    subgraph3.node('stupid1','Cory')
    subgraph3.node('stupid2','Trevor')
    subgraph.attr(rank='same')

tree.edge('park_owner','supervisor',arrowhead="none")
tree.edge('park_owner','supervisor2',arrowhead="none")

tree.edge('supervisor','junky1',arrowhead="none")
tree.edge('supervisor','junky2',arrowhead="none")
tree.edge('supervisor','junky3',arrowhead="none")

tree.edge('supervisor2','junky1',arrowhead="none")
tree.edge('supervisor2','junky2',arrowhead="none")
tree.edge('supervisor2','junky3',arrowhead="none")

tree.edge('junky1','stupid1',arrowhead="none")
tree.edge('junky1','stupid2',arrowhead="none")

tree
```

Drzewo genealogiczne przedstawia relacje między postaciami w serialu "Chłopaki z baraku"
